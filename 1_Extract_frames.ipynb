```python
# This notebook saves each video frame as an image.
import cv2
import os

# Set the video file path.
input_video_path = r"input_video\video.mp4"
# Make the frame output folder if needed.
if not os.path.exists("extract_frame"):
    os.makedirs("extract_frame")

# Open the video for reading.
cap = cv2.VideoCapture(input_video_path)
frame_count = 0

# Read and save frames until the video ends.
while True:
    ret, frame = cap.read()
    if not ret:
        break
    # Create a clear file name for this frame.
    frame_name = os.path.join("extract_frame", f"frame_{frame_count:05d}.jpg")
    cv2.imwrite(frame_name, frame)
    frame_count += 1
# Close the video file.
cap.release()
print(f"Extracted {frame_count} frames.")
```
